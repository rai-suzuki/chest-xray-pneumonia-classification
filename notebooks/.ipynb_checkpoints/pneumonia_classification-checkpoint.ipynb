# 胸部X線画像による肺炎分類

　胸部X線画像を使って、肺炎の有無を分類する画像分類モデルを作成します。

　題材は **Kaggle** の **Chest X-Ray Images (Pneumonia)** を使います。

　これは広州の女性児童医療センターの患者から収集された1〜5歳の小児の胸部X線レントゲン画像です。

　ここでは本分析に使用するデータやライブラリ、ハードウェアの確認などを行います。

#### 使用データ

#### 分類クラス
- `NORMAL`: 肺炎なし
- `PNEUMONIA`: 肺炎あり

#### ⑷ グレースケールかRGBか
　画像データについては、グレースケールは白か黒（0から255）の値１つで表されるのに対し、RGBは赤、緑、青（それぞれ0から255）の値３つで表されることから、データのスケールについては確認が必要です。注意してほしいのは、RGB画像でも３つの値それぞれが全て同じ値であるときはグレースケール（白か黒）で表されることです。今回のtrainデータのRGB画像は、全てグレースケールで表されています（カラー画像はなし）。

```python
mode_data = []

for phase in phases:
    for cls in ['NORMAL', 'PNEUMONIA']:
        dir_path = os.path.join(data, phase, cls)
        img_paths = glob.glob(os.path.join(dir_path, '*.jpeg')) 
            
        for path in img_paths:
            with Image.open(path) as img:
                mode_data.append({
                    'Phase': phase,
                    'Class': cls,
                    'Mode' : img.mode
                })

df_modes = pd.DataFrame(mode_data)

mode_counts = pd.crosstab(df_modes['Phase'], df_modes['Mode'])
mode_counts = mode_counts.reindex(['train', 'val', 'test'])
mode_counts
```

#### ⑺ 破損画像や読み込み失敗の有無
　ここでは、扱うKaggleのデータで使用することができない画像データがないかを確認します。実際にはまず、画像ファイルの型（ヘッダー）を確認し、そのファイルの中身が画像ファイルとして適した形になっているかを確認します。これら二つの過程でエラーを検知したデータを破損や読み込みの失敗として扱います。

```python
corrupted_files = []
    
for phase in phases:
    for cls in classes:
        dir_path = os.path.join(data, phase, cls)
        img_paths = glob.glob(os.path.join(dir_path, '*.*'))

        for path in img_paths:
            try:
                with Image.open(path) as img:
                    img.verify()
            except Exception as e:
                corrupted_files.append((path, str(e)))
                
if len(corrupted_files) == 0:
    print(f"破損画像や読み込みの失敗はなし")
else:
    print(f"破損画像や読み込みの失敗: {corrupted_files} 件")
```

### 2. 前処理とデータの拡張

　前処理は、異なる画像サイズを統一し、色味を整えることでAIが計算しやすい形にするためにします。

　また、データの拡張を行うことで、AIの丸暗記を防いで未知の画像への対応力を高めることができます。

　データの前処理については以下のとおりです。
- リサイズ: 224x224
- Tensor化
- 正規化

　また、学習時におけるデータの拡張については以下のとおりです。
- 軽い回転: -10度 から +10度
- 軽い平行移動
- 軽い拡大縮小
- 明るさ、コントラストの微調整
- 左右反転

```python
image_size = 224
mean = [0.485, 0.456, 0.406]
std = [0.229, 0.224, 0.225]

# 前処理（評価用）
eval_transforms = transforms.Compose([
    transforms.Resize((image_size, image_size)),
    transforms.ToTensor(),
    transforms.Normalize(mean, std)
])

# 前処理　（訓練用）
train_transforms = transforms.Compose([
    transforms.Resize((image_size, image_size)),
    
    # データ拡張  ------------------------------------
    transforms.RandomRotation(degrees=10),
    transforms.RandomAffine(degrees=0,
                            translate=(0.05, 0.05),
                            scale=(0.95, 1.05)),
    transforms.ColorJitter(brightness=0.1,
                           contrast=0.1),
    transforms.RandomHorizontalFlip(p=0.5),
    # ----------------------------------------------
    
    transforms.ToTensor(),
    transforms.Normalize(mean, std)
])

# 前処理データの整理
data_transforms = {
    'train': train_transforms,
    'val': eval_transforms,
    'test': eval_transforms
}
```

### 3. モデル実装

　今回検証していくモデルは以下のとおりです。
- ⑴ Baseline CNN
- ⑵ ResNet18
- ⑶ EfficientNet

#### ⑴ Baseline CNN
　ベースラインを作ることで、他モデルとの比較対象とします。

```python
class CNN(nn.Module):
    def __init__(self, num_classes=2):
        super().__init__()
        self.features = nn.Sequential(
            nn.Conv2d(3, 16, 3, padding=1), nn.ReLU(), nn.MaxPool2d(2),
            nn.Conv2d(16, 32, 3, padding=1), nn.ReLU(), nn.MaxPool2d(2),
        )
        self.classifier = nn.Sequential(
            nn.Linear(32 * 56 * 56, 128), nn.ReLU(), nn.Dropout(0.5),
            nn.Linear(128, num_classes)
        )
    def forward(self, x):
        x = self.features(x)
        x = torch.flatten(x, 1)
        return self.classifier(x)

model_cnn = CNN().to(device)
optimizer_cnn = optim.Adam(model_cnn.classifier.parameters(), lr=1e-3)
```

#### ⑵ ResNet18
　最初の転移学習モデルとして採用しています。構造を理解しやすく、学習も比較的軽いため扱いやすいモデルです。

```python
model_resnet = models.resnet18(weights=models.ResNet18_Weights.IMAGENET1K_V1)
for param in model_resnet.parameters():
    param.requires_grad = False
model_resnet.fc = nn.Linear(model_resnet.fc.in_features, 2)
model_resnet = model_resnet.to(device)

optimizer_resnet = optim.Adam(model_resnet.fc.parameters(), lr=1e-3)
```

#### ⑵【発展】 ResNet18(Fine-tuning)
　Recallに特化して検証します。

```python
model_ft = models.resnet18(weights=models.ResNet18_Weights.IMAGENET1K_V1)
for param in model_ft.parameters():
    param.requires_grad = False

num_ftrs = model_ft.fc.in_features
model_ft.fc = nn.Linear(num_ftrs, 2)

for name, child in model_ft.named_children():
    if name in ['layer3', 'layer4']:
        for param in child.parameters():
            param.requires_grad = True

model_ft = model_ft.to(device)

optimizer_ft = optim.Adam([
    {'params': model_ft.layer3.parameters(), 'lr': 1e-4},
    {'params': model_ft.layer4.parameters(), 'lr': 1e-4},
    {'params': model_ft.fc.parameters(), 'lr': 1e-3}
])
```

#### ⑶ EfficientNet-B0
　比較的軽量で性能も出やすいモデルです。

```python
#6 EfficientNet-B0(転移学習)
model_effnet = models.efficientnet_b0(weights=models.EfficientNet_B0_Weights.IMAGENET1K_V1)
for param in model_effnet.parameters():
    param.requires_grad = False
model_effnet.classifier[1] = nn.Linear(model_effnet.classifier[1].in_features, 2)
model_effnet = model_effnet.to(device)

optimizer_effnet = optim.Adam(model_effnet.classifier.parameters(), lr=1e-3)
```

### 4. 学習

　学習における各項目の設定については以下のとおりです。
- 入力サイズ：224×224
- バッチサイズ：32
- optimizer：Adam
- 学習率：1e-3
- エポック数：5
- 損失関数：CrossEntropyLoss

　なお、ResNet18については、Fine-tuningを行った場合も合わせて実装しています。

#### ⑴ 学習の設定

```python
# データセットとデータローダーの作成
image_datasets = {x: datasets.ImageFolder(os.path.join(data, x),
                                          data_transforms[x]) for x in phases}

batch_size = 32
dataloaders = {x: DataLoader(image_datasets[x],
                             batch_size=batch_size,
                             shuffle=(x == 'train'),
                             num_workers=2) for x in phases}
dataset_sizes = {x: len(image_datasets[x]) for x in phases}
class_names = image_datasets['train'].classes

num_epochs = 1
criterion = nn.CrossEntropyLoss()

# ※将来的に追加する場合のイメージ（今は実行しなくてOKです）
# 正常(0)に対するペナルティを3倍、肺炎(1)に対するペナルティを1倍にする
# weights = torch.tensor([3.0, 1.0]).to(device)
# criterion = nn.CrossEntropyLoss(weight=weights)

def train_model(model, criterion, optimizer, num_epochs):
    since = time.time()
    best_model_wts = copy.deepcopy(model.state_dict())
    best_acc = 0.0

    for epoch in range(num_epochs):
        print(f'Epoch {epoch+1}/{num_epochs}')

        for phase in ['train', 'val']:
            if phase == 'train':
                model.train()  # 学習モード
            else:
                model.eval()   # 評価モード

            running_loss = 0.0
            running_corrects = 0

            for inputs, labels in dataloaders[phase]:
                inputs = inputs.to(device)
                labels = labels.to(device)

                optimizer.zero_grad()

                with torch.set_grad_enabled(phase == 'train'):
                    outputs = model(inputs)
                    _, preds = torch.max(outputs, 1)
                    loss = criterion(outputs, labels)

                    if phase == 'train':
                        loss.backward()
                        optimizer.step()

                running_loss += loss.item() * inputs.size(0)
                running_corrects += torch.sum(preds == labels.data)

            epoch_loss = running_loss / dataset_sizes[phase]
            epoch_acc = running_corrects.double() / dataset_sizes[phase]

            print(f'{phase} Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}')

            if phase == 'val' and epoch_acc > best_acc:
                best_acc = epoch_acc
                best_model_wts = copy.deepcopy(model.state_dict())

    time_elapsed = time.time() - since
    print(f'Training complete in {time_elapsed // 60:.0f}m {time_elapsed % 60:.0f}s')
    print(f'Best val Acc: {best_acc:4f}')

    model.load_state_dict(best_model_wts)
    return model
```

#### ⑵ 学習の実行

```python
model_cnn = train_model(model_cnn, criterion, optimizer_cnn, num_epochs=num_epochs)
```

```python
model_resnet = train_model(model_resnet, criterion, optimizer_resnet, num_epochs=num_epochs)
```

```python
model_ft = train_model(model_ft, criterion, optimizer_ft, num_epochs=num_epochs)
```

```python
model_effnet = train_model(model_effnet, criterion, optimizer_effnet, num_epochs=num_epochs)
```

### 5. 評価指標

　このタスクでは、Accuracyだけでモデルを評価するのは不十分です。肺炎を見逃すこと、つまり偽陰性を減らす（Recall）観点が重要です。評価指標は以下のとおりです。
- Accuracy：全体の正解率を見る指標
- Precision：実際に肺炎ではない画像を、どれだけ肺炎として検出したかを見る指標（誤検出）
- Recall：実際に肺炎である画像を、どれだけ肺炎として検出できたかを見る指標（見逃し）
- F1-score：PrecisionとRecallのバランスの良さを見る指標
- Confusion Matrix：実際に肺炎か否かと予測の結果を表に集計したもの

```python
def evaluate(model, dataloader):
    model.eval()
    all_preds = []
    all_labels = []
    
    with torch.no_grad():
        for inputs, labels in dataloader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
            
    print(f"Accuracy : {accuracy_score(all_labels, all_preds):.4f}")
    print(f"Precision: {precision_score(all_labels, all_preds):.4f}")
    print(f"Recall   : {recall_score(all_labels, all_preds):.4f}")
    print(f"F1-score : {f1_score(all_labels, all_preds):.4f}")
    
    cm = confusion_matrix(all_labels, all_preds)
    sns.heatmap(cm,
                annot=True,
                fmt='d',
                cmap='Blues',
                xticklabels=class_names,
                yticklabels=class_names)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()
    
    return all_preds, all_labels
```

```python
# CNN
evaluate(model_cnn, dataloaders['test']);
```

```python
# ResNet18
evaluate(model_resnet, dataloaders['test']);
```

```python
evaluate(model_ft, dataloaders['test']);
```

```python
# EfficientNet-B0
evaluate(model_effnet, dataloaders['test']);
```

### 6. 誤分類分析

```python
def analyze_misclassifications(model, dataloader, device, class_names, mean, std, num_images=4):
    model.eval()
    all_errors = []
    
    with torch.no_grad():
        for inputs, labels in dataloader:
            inputs = inputs.to(device)
            labels = labels.to(device)
            
            outputs = model(inputs)
            probs = torch.softmax(outputs, dim=1) # 確率を計算
            _, preds = torch.max(outputs, 1)
            
            for i in range(inputs.size(0)):
                true_label = labels[i].item()
                pred_label = preds[i].item()
                
                # 間違えた画像だけを記録
                if true_label != pred_label:
                    confidence = probs[i, pred_label].item() 
                    all_errors.append({
                        'image_tensor': inputs[i].cpu(),
                        'true_lbl': true_label,
                        'pred_lbl': pred_label,
                        'true_name': class_names[true_label],
                        'pred_name': class_names[pred_label],
                        'confidence': confidence
                    })

    if not all_errors:
        print("誤分類された画像はありません。")
        return

    # ==========================================
    # 4つの条件でリストを分類・ソート
    # ==========================================
    # 1. NORMAL -> PNEUMONIA (誤報)
    fp_errors = [e for e in all_errors if e['true_name'] == 'NORMAL' and e['pred_name'] == 'PNEUMONIA']
    
    # 2. PNEUMONIA -> NORMAL (見逃し・クリティカル)
    fn_errors = [e for e in all_errors if e['true_name'] == 'PNEUMONIA' and e['pred_name'] == 'NORMAL']
    
    # 3. 予測確率が高い（自信満々）順
    high_conf_errors = sorted(all_errors, key=lambda x: x['confidence'], reverse=True)
    
    # 4. 予測確率が低い（迷っている）順：2クラス分類なので0.5に近いもの
    low_conf_errors = sorted(all_errors, key=lambda x: x['confidence'])

    # ==========================================
    # 描画用の関数
    # ==========================================
    def show_images(results, title, num):
        show_list = results[:num]
        if not show_list:
            print(f"[{title}] に該当する画像はありません。")
            return
            
        fig, axes = plt.subplots(1, len(show_list), figsize=(4 * len(show_list), 4))
        if len(show_list) == 1: axes = [axes]
            
        for i, res in enumerate(show_list):
            img = res['image_tensor'].numpy().transpose((1, 2, 0))
            img = np.array(std) * img + np.array(mean)
            img = np.clip(img, 0, 1)
            
            axes[i].imshow(img)
            
        plt.suptitle(title, fontsize=14, y=1.05)
        plt.tight_layout()
        plt.show()

    print(f"総誤分類数: {len(all_errors)} 枚")
    show_images(fp_errors, "1. NORMAL を PNEUMONIA と誤分類 (過剰検出)", num_images)
    show_images(fn_errors, "2. PNEUMONIA を NORMAL と誤分類 (見逃し)", num_images)
    show_images(high_conf_errors, "3. 予測確率が高いのに間違えた例 (自信満々のミス)", num_images)
    show_images(low_conf_errors, "4. 予測確率が低く迷っている例 (境界線の画像)", num_images)
```

```python
analyze_misclassifications(model_cnn, dataloaders['test'], device, class_names, mean, std, num_images=4)
```

```python
analyze_misclassifications(model_resnet, dataloaders['test'], device, class_names, mean, std, num_images=4)
```

```python
analyze_misclassifications(model_effnet, dataloaders['test'], device, class_names, mean, std, num_images=4)
```

　※Grad-camはresnetにのみ適用。

```python
from pytorch_grad_cam import GradCAM
from pytorch_grad_cam.utils.image import show_cam_on_image
from pytorch_grad_cam.utils.model_targets import ClassifierOutputTarget

def visualize_gradcam(model, image_path, true_label_name):
    print(f"[{true_label_name}] の画像に対する Grad-CAM を生成します...")
    
    # ==================================================
    # ★追加：Grad-CAMが勾配（根拠）を逆算できるように、
    # 一時的にモデル全体の凍結を解除（Unfreeze）する
    # ==================================================
    for param in model.parameters():
        param.requires_grad = True

    # 1. 画像の読み込みと前処理（Step 2と同じ設定）
    img_pil = Image.open(image_path).convert('RGB')
    mean, std = [0.485, 0.456, 0.406], [0.229, 0.224, 0.225]
    
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean, std)
    ])
    input_tensor = transform(img_pil).unsqueeze(0).to(device) # バッチ次元を追加
    
    # 2. 表示用の元の画像（正規化されていない、0〜1の範囲のRGB）を作成
    img_resized = img_pil.resize((224, 224))
    img_display = np.float32(img_resized) / 255.0

    # 3. AIに予測させる（ここは確率を出すだけなので勾配不要）
    model.eval()
    with torch.no_grad():
        output = model(input_tensor)
        prob = torch.softmax(output, dim=1)[0, 1].item() # PNEUMONIA(1)の確率
        pred_idx = torch.argmax(output, dim=1).item()
        pred_label_name = 'PNEUMONIA' if pred_idx == 1 else 'NORMAL'

    # 4. Grad-CAMのセットアップ
    target_layers = [model.layer4[-1]]
    cam = GradCAM(model=model, target_layers=target_layers)
    targets = [ClassifierOutputTarget(pred_idx)]
    
    # 5. ヒートマップの生成（ここで勾配計算が行われるので、上のUnfreezeが効きます！）
    grayscale_cam = cam(input_tensor=input_tensor, targets=targets)[0, :]
    
    # 元の画像に、ヒートマップを重ね合わせる
    visualization = show_cam_on_image(img_display, grayscale_cam, use_rgb=True)

    # 6. 結果の描画
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    axes[0].imshow(img_display)
    axes[0].set_title(f"Original Image\nTrue: {true_label_name}")
    axes[0].axis('off')
    
    axes[1].imshow(visualization)
    axes[1].set_title(f"Grad-CAM\nPred: {pred_label_name} (Prob: {prob:.2f})")
    axes[1].axis('off')
    
    plt.tight_layout()
    plt.show()

# 実行例
sample_pneumonia_path = './chest_xray/test/PNEUMONIA/person1_virus_7.jpeg' # 任意の画像を選択
visualize_gradcam(model_resnet, sample_pneumonia_path, "PNEUMONIA")
```

### 7. 推論速度とモデルサイズ比較

```python
import time
import os
import tempfile
import torch

def measure_real_inference(model, dataloader, model_name):
    # 現在の環境がGPUかCPUかを自動判定
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model = model.to(device)
    model.eval()

    print(f"[{model_name}] の性能を {device.type.upper()} 環境で計測中...")

    # ==========================================
    # 1. モデルファイルサイズ (MB) の計測
    # ==========================================
    with tempfile.NamedTemporaryFile(delete=False) as f:
        # 重みを一時ファイルに保存して実際の容量を測る
        torch.save(model.state_dict(), f.name)
        file_size_mb = os.path.getsize(f.name) / (1024 * 1024)
    os.remove(f.name) # 計測後は削除

    # ==========================================
    # 2. 実際のテストデータを用いた推論速度の計測
    # ==========================================
    total_time = 0.0
    total_images = 0

    # 計測を正確にするためのウォームアップ（最初の1バッチだけ空回しする）
    # ※GPUなどは最初の処理に初期化の時間がかかるため
    warmup_inputs, _ = next(iter(dataloader))
    warmup_inputs = warmup_inputs.to(device)
    with torch.no_grad():
        _ = model(warmup_inputs)

    # 本番計測
    with torch.no_grad():
        for inputs, _ in dataloader:
            inputs = inputs.to(device)
            batch_size = inputs.size(0)
            
            # GPUの場合は非同期処理のズレを防ぐため、処理が完了するまで待機（同期）させる
            if device.type == 'cuda':
                torch.cuda.synchronize()
            
            # バッチごとの計測開始
            start_time = time.time()
            _ = model(inputs)
            
            if device.type == 'cuda':
                torch.cuda.synchronize()
            # 計測終了
            end_time = time.time()
            
            total_time += (end_time - start_time)
            total_images += batch_size

    # 1枚あたりの平均推論時間 (ミリ秒)
    avg_latency = (total_time / total_images) * 1000

    print(f" - モデルサイズ : {file_size_mb:.2f} MB")
    print(f" - 推論時間     : {avg_latency:.2f} ms/image\n")

    return file_size_mb, avg_latency
```

```python
measure_real_inference(model_cnn, dataloaders['test'], "Baseline CNN")
```

```python
measure_real_inference(model_resnet, dataloaders['test'], "ResNet18")
```

```python
measure_real_inference(model_effnet, dataloaders['test'], "EfficientNet-B0")
```
